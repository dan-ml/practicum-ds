# Анализ международного рынка компьютерных игр 

В проекте требуется проанализировать исторические данные о продажах, оценках, жанрах компьютерных игр до 2016 г. Цель - выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Проект должен включать шаги: изучение данных из источника, подготовка и предобработка данных, исследовательский анализ, формулировка и проверка гипотез, формулировка выводов.

Загрузим данные из файла и ознакомимся с таблицей.

```python
try:
    main_data = pd.read_csv('games.csv')
except:
    main_data = pd.read_csv('/datasets/games.csv')
main_data.info()
```

```python
main_data.head(10)
```

Столбцы в таблице не приведены к нижнему регистру. В некоторых столбцах имеются пропуски. Эти проблемы будем исправлять в п. 2.

## Подготовка и предобработка данных

### Переименование столбцов

Приведем названия столбцов к нижнему регистру.

```python
main_data.columns = main_data.columns.str.lower()
main_data.columns
```

Переименование столбцов прошло успешно.

### Преобразование типов данных

Преобразуем типы данных в столбцах. Сначала изменим тип данных столбца year_of_release на целочисленный Int64, допускающий наличие пропусков.

```python
main_data['year_of_release'] = main_data['year_of_release'].astype('Int64')
main_data['year_of_release'].head(5)
```

Столбец user_score имеет тип object, а должен иметь тип float64. При преобразование строки в вещественное число могут быть ошибки преобразования. Посмотрим на уникальные значения столбца.

```python
main_data['user_score'].unique()
```

Значение tbd не сможет преобразоваться в вещественный тип данных. Данное значение для рейтинга расшифровывается как "будет определено". Т.е. если игра ещё не вышла на рынок, то пользователи не успели её оценить, но при этом игра уже была оценена критиками, имеющими доступ к предрелизной версии. Но следует посмотреть, какие данные содержатся в таблице у игр с рейтингом tbd.

```python
main_data.loc[main_data['user_score']=='tbd']
```

Значения tbd имеют в том числе и игры, выпущенные намного раньше 2016 года. Т.е. теория о свежих играх, которые пользователи не успели оценить, отпадает. Вероятно, при составлении таблицы значения tbd были выставлены даже тем играм, которые официально пользователями не оценивались таким типом рейтинга. На момент выхода игры такого рейтинга могло не существовать и база данных по нему не велась. Поэтому заменим значения tbd на пропуски (np.nan). После этого выполним преобразование типа столбца.

```python
main_data.loc[main_data['user_score']=='tbd', 'user_score'] = np.nan
```

```python
main_data['user_score'] = main_data['user_score'].astype('float64')
main_data.info()
```

Преобразование типа столбца user_score прошло успешно. 

### Заполнение пропусков

Теперь стоит попытаться заполнить пропуски в различных столбцах. Первый столбец с пропусками - name. В этом столбце всего два пропуска. Значения столбца очень важны, а пропусков очень мало относительно общего числа строк. Поэтому удалим строки с этими пропусками.

```python
main_data = main_data.dropna(subset=['name']).reset_index(drop=True)
main_data.info()
```

Количество строк в таблице уменьшилось на две. Одновременно удалились строки с пропусками в столбце genre.

Следующий столбец с пропусками - year_of_release. Оставим пропуски в этом столбце как есть, так как их нельзя заполнять нулями или усреднёнными значениями. Причины наличия таких пропусков: плохая подготовка исходных данных.

Следующие столбцы с пропусками: critic_score, user_score и rating. Заполнять эти пропуски тоже не будем. Часть пропусков может быть обусловлена плохой подоготовкой исходных данных, а часть пропусков просто свидетельствует о том, что рассматриваемая игра не оценивалась критиками и пользователями (конкретно этих оценок ещё не существовало на момент выхода игры). 

Далее рассмотрим столбец с пропусками rating. На момент выхода игры мог ещё не существовать рейтинг от организации ESRB. Также для определённой доли игр рейтинг мог вообще не присваиваться. Поэтому столбец содержит много пропусков. Так как значения столбца категориальные, то заменим пропуски на значение *Unknown*. Значения этого столбца будут использованы при анализе в п. 4 проекта.

```python
main_data['rating'].fillna('Unknown', inplace=True)
main_data.info()
```

## Исследовательский анализ данных

### Выпуск игр по годам

```python
(main_data.pivot_table(index='year_of_release', values='name', aggfunc='count').
 plot(title='Выпуск игр по годам (все платформы)', kind='line', figsize=(14,5), grid=True,
                        xlabel='Годы', ylabel='Количество игр'))
plt.minorticks_on()
plt.grid(which='minor', linestyle = ':')
plt.show()
```

Больше всего игр выпускалось в 2008-2009 годах (более 1400 игр в год).С 2000-го года наблюдаются два этапа роста выпуска новых игр. Вероятно, причина кроется в появлении новых игровых приставок с красивой для того времени 3d-графикой: Sega DreamCast и Sony Playstation 2 (чуть позднее - ещё и Xbox). Также происходило развитие портативных консолей. Данные за 2016 год неполные, поэтому виден спад на графике для этого года.

Построим графики выпуска игр по отдельным платформам в период с 2004-го года (начало второго этапа выпуска игр). Это поможет определить, какие платформы ещё актуальны для рынка, и на них выпускаются игры. Воспользуемся сводной таблице с группировкой по игровой платформе и со столбцами, равными годам выпуска игры. После формирования сводную таблицу следует транспонировать, чтобы можно было затем построить сразу несколько графиков по столбцам. Так как платформ много, то разобъём таблицу на две части и построим два графика: с первой частью столбцов и со второй.

```python
fig1, ax1 = plt.subplots(figsize=(15,8))
(main_data.query('year_of_release>=2004').
 pivot_table(index='platform', columns='year_of_release', values='name', aggfunc='count').T.
 loc[:,'3DS':'PS3'].plot(ax=ax1))
ax1.set_title('Выпуск игр по годам (раздельно по платформам, часть 1)')
ax1.set_xlabel('Годы')
ax1.set_ylabel('Количество игр')
ax1.legend(loc='upper left')
plt.minorticks_on()
ax1.grid(which='major', linestyle='-')
ax1.grid(which='minor', linestyle=':')
plt.show()

fig1, ax1 = plt.subplots(figsize=(15,8))
(main_data.query('year_of_release>=2004').
 pivot_table(index='platform', columns='year_of_release', values='name', aggfunc='count').T.
 loc[:,'PS4':'Xone'].plot(ax=ax1))
ax1.set_title('Выпуск игр по годам (раздельно по платформам, часть 2)')
ax1.set_xlabel('Годы')
ax1.set_ylabel('Количество игр')
ax1.legend(loc='upper left')
plt.minorticks_on()
ax1.grid(which='major', linestyle='-')
ax1.grid(which='minor', linestyle=':')
plt.show()
```

**Вывод:** лидеры по выпущенным играм - портативные платформы Nindendo DS и Nindendo Wii. Но эти игровые платформы уже практически закончили свой жизненный цикл к 2016 г. Из всех платформ в плане выпуска новых игр на момент 2016 г. логично рассматривать PC, WiiU, PlayStation 4 и Xbox One. Также допустимо рассматривать портативные консоли Nindendo 3DS и PlayStation Vita, но выпуск игр на них активно снижается. Данные до 2011 г. уже неактуальны, так как платформы до этого года уже закончили или заканчивают свой жизненный цикл на 2016 г.

### Изменение продаж игр по платформам

Создадим сводную таблицу, затем отсортируем её для определения платформ с наибольшими суммарными продажами.

```python
data_total_sales = (main_data.pivot_table(index='platform', values='total_sales', aggfunc='sum').
                    sort_values(by='total_sales', ascending=False))
(data_total_sales.plot(title='Продажи игр по платформам', kind='line', figsize=(16,5), grid=True, 
                       xlabel='Название платформы', ylabel='Объём продаж (млн)', fontsize='8'))
plt.xticks(np.arange(len(data_total_sales)), labels = data_total_sales.index.values)
plt.show()
```

Исходя из графика по объёмам продаж резко выделяются первые шесть платформ (PS2, X360, PS3, Wii, DS, PS). Для этих платформ и построим распределение по годам. Создадим ещё одну сводную таблицу с группировкой по платформе и году. Затем из этой таблицы выполним срез данных, в который войдут только платформы из заранее созданного списка. После этого транспонируем таблицу, чтобы построить сразу несколько графиков по столбцам.

```python
platform_list = ['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS']
fig1, ax1 = plt.subplots(figsize=(15,6))
(main_data.pivot_table(index='platform', columns='year_of_release', values='total_sales', aggfunc='sum').
 query('platform in @platform_list')).T.plot(ax=ax1)
ax1.set_title('Продажи игр по годам (раздельно по платформам)')
ax1.set_xlabel('Год выхода игр')
ax1.set_ylabel('Суммарные продажи (млн копий)')
plt.minorticks_on()
ax1.grid(which='major', linestyle='-')
ax1.grid(which='minor', linestyle=':')
plt.show()
```

**Вывод:** по графикам на примере платформ с наибольшими суммарными продажами игр можно определить, что средний срок жизни игровых платформ - **10 лет**. За это время игровая платформа выходит на рынок, начинают расти продажи игр для неё, затем в конце периода продажи прекращаются. Этот момент можно считать исчезанием игровой платформы с рынка.

### Выбор актуального периода

На основе выводов в пп. 3.1 и 3.2 для дальнейшего исследования в качестве **актуального периода** выберем **период с 2013 г**. Платформы, игры на которые активно продавались и выпускались до 2013 г, уже закончат или будут близки к завершению жизненного цикла в 2017 году. И нелогично на основе этих данных строить прогноз. Также ограничимся рассмотрением следующих игровых платформ, которые могут влиять на игровой рынок в 2017 году: **PC, 3DS, WiiU, PS4, XOne, PSV**.

Сразу создадим срез данных *main_data_act* по указанным параметрам и в дальнейшем будем работать с ним.

```python
actual_year= 2013
platform_list_act = ['PC','3DS','WiiU','PS4','XOne','PSV']
main_data_act = main_data.query('year_of_release>=@actual_year and platform in @platform_list_act').reset_index(drop=True)
main_data_act.head(5)
```

### Выбор потенциально прибыльных платформ

### Влияние жанров игр на продажи

## Проверка гипотез

### Гипотеза 1

Требуется проверить, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Сформулируем нулевую и альтернативную гипотезы:
- нулевая гипотеза: средние пользовательские рейтинги игр на платформах Xbox One и PC равны;
- альтернативная гипотеза: средние пользовательские рейтинги игр на платформах Xbox One и PC различаются.

Для проверки гипотезы воспользуемся t-критерием Стьюдента и применим метод `ttest_ind`. Все выборки содержат более 30 элементов. Параметр equal_var установим в значение False, так как выборки имеют разный размер (соответственно могут различаться дисперсии). 

В вызове метода `ttest_ind` сформируем необходимые срезы данных по платформам, а также избавимся от пропусков в них.

```python
alpha = 0.05
results = (st.ttest_ind(main_data_act.query('platform=="XOne"')['user_score'].dropna(), 
                        main_data_act.query('platform=="PC"')['user_score'].dropna(), equal_var=False))
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод**: нулевая гипотеза о том, средние пользовательские рейтинги игр на платформах Xbox One и PC равны, не отвергается.

### Гипотеза 2

Требуется проверить, что cредние пользовательские рейтинги жанров Action и Sports разные. Сформулируем нулевую и альтернативную гипотезы:
- нулевая гипотеза: cредние пользовательские рейтинги игр жанров Action и Sports равны;
- альтернативная гипотеза: cредние пользовательские рейтинги игр жанров Action и Sports различаются.

Для проверки гипотезы воспользуемся t-критерием Стьюдента и применим метод `ttest_ind`. Все выборки содержат более 30 элементов. Параметр equal_var установим в значение False, так как выборки имеют разный размер (соответственно могут различаться дисперсии). 

В вызове метода `ttest_ind` сформируем необходимые срезы данных по жанрам.

```python
alpha = 0.05
results = (st.ttest_ind(main_data_act.query('genre=="Action"')['user_score'].dropna(), 
                        main_data_act.query('genre=="Sports"')['user_score'].dropna(), equal_var=False))
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод**: нулевая гипотеза о том, что cредние пользовательские рейтинги игр жанров Action и Sports равны, отвергается в пользу альтернативной. А именно, cредние пользовательские рейтинги игр жанров Action и Sports различаются.
