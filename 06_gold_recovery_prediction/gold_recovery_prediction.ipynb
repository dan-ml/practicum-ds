### Проверка значений эффективности обогащения руды

Проверим правильность значений эффективности обогащения чернового концентрата (rougher.output.recovery), которые находятся в обучающей выборке и являются одним из целевых признаков. Сравним эти значения с числами, полученными в результате расчёта по формуле, и посчитаем метрику MAE c использованием функции mean_absolute_error.

```python
columns_ROR = data_train[['rougher.input.feed_au','rougher.output.concentrate_au','rougher.output.tail_au',
                          'rougher.output.recovery']]
columns_ROR.info()
```

В необходимых для расчётов столбцах нет значений NaN. Можно приступать к расчёту.

```python
rougher_tail = columns_ROR['rougher.output.tail_au']
rougher_recovery = (columns_ROR['rougher.output.concentrate_au']*(columns_ROR['rougher.input.feed_au']-rougher_tail)/
         (columns_ROR['rougher.input.feed_au']*(columns_ROR['rougher.output.concentrate_au']-rougher_tail)))*100

print(f'Значение MAE для rougher.output.recovery: {mean_absolute_error(columns_ROR["rougher.output.recovery"], rougher_recovery):.3}')
```

**Вывод:** значение MAE пренебрежимо мало, поэтому будет считать, что в обучающей выборке содержатся верные значения эффективности обогащения чернового концентрата.

## Анализ данных

### Анализ изменения концентрации металлов на различных этапах очистки

Проанализируем изменение концентрации металлов (золото Au, серебро Ag, свинец Pb) на различных этапах очистки. В полном датасете доступно три группы сведений о долях металлов в концентратах: rougher.output.concentrate_XX, primary_cleaner.output.concentrate_XX, final.output.concentrate_XX. ## Построение моделей

### Функции для расчёта метрик качества модели

Напишем функции для расчёта метрик качества модели, которые в дальнейшем будем использовать. В одной из функций будет использоваться техника кросс-валидации модели.

Напишем функцию для расчёта метрики sMAPE (Symmetric Mean Absolute Percentage Error). Она будет вычислять метрику для ОДНОГО целевого признака, поданного на вход в качестве аргумента.

```python
def sMAPE(target, predictions):
    return sum(abs(target - predictions)/abs((target+predictions)/2))/len(target)*100
```

В процессе работы над проектом выяснилось, что для работы функции GridSearchCV, будет необходима отдельная функция расчёта метрики sMAPE, которая будет работать сразу с ДВУМЯ целевыми признаками, поданными на вход в качестве аргумента.

```python
def sMAPE_2d(target_2d, predictions_2d):
    sMAPE_rougher=(sum(abs(target_2d['rougher.output.recovery']-predictions_2d[:,0])/
                      abs((target_2d['rougher.output.recovery']+predictions_2d[:,0])/2))/
                   len(target_2d['rougher.output.recovery'])*100)
    
    sMAPE_final=(sum(abs(target_2d['final.output.recovery']-predictions_2d[:,1])/
                      abs((target_2d['final.output.recovery']+predictions_2d[:,1])/2))/
                   len(target_2d['final.output.recovery'])*100)   
    
    return 0.25*sMAPE_rougher + 0.75*sMAPE_final
```

Напишем функцию оценки качества модели с помощью кросс-валидации (непосредственно для оценки качества будет применяться написанная ранее функция sMAPE). Функция будет принимать в качестве аргумента target только одномерную структуру данных (к примеру, Series). При кросс-валидации будет вычисляться не метрика по умолчанию, а будет вызываться созданная выше функция *sMAPE*. Чем меньше значение этой метрики, тем лучше. За это отвечает параметр *greater_is_better=False*. При указании этого параметра значение метрики умышленно становится отрицательным, поэтому при выводе результата функции *my_cross_val* используем значение по модулю.

```python
def my_cross_val(estimator, features, target):
    scores = cross_val_score(estimator, features, target, cv=5, scoring=make_scorer(sMAPE, greater_is_better=False))
    final_score = np.mean(scores)
    return abs(final_score)
```

Теперь создадим функцию для расчёта итогового значения метрики sMAPE (с учётом обоих целевых признаков rougher.output.recovery и final.output.recovery) и вывода результатов. Так как функция использует технику кросс-валидации, то в качестве аргументов в эту функцию следует подавать признаки из обучающей выборки. Функция принимает в качестве аргумента target двухмерную структуру данных (в нашем случае - датафрейм с двумя столбцами).

```python
def model_result(estimator, features, target_2d):
    sMAPE_rougher = my_cross_val(estimator, features, target_2d['rougher.output.recovery']) 
    sMAPE_final = my_cross_val(estimator, features, target_2d['final.output.recovery'])
    sMAPE_total = 0.25*sMAPE_rougher + 0.75*sMAPE_final
    print(f'Значение метрики sMAPE для предсказаний величины rougher.output.recovery: {sMAPE_rougher:.2f} %')
    print(f'Значение метрики sMAPE для предсказаний величины final.output.recovery: {sMAPE_final:.2f} %')
    print(f'Значение итоговой метрики sMAPE (с учётом эффективностей обогащения на двух этапах): {sMAPE_total:.2f} %')
```

Далее проведена предобработка данных: выполнены преобразования типов данных, заполнены пропуски во всех датасетах, изменены индексы. Признаки, которые отсутствовали в тестовой выборке, удалены из обучающей выборки. В тестовую выборку добавлены целевые признаки из полного датасета.

Выполнен исследовательский анализ данных:
- проанализировано изменение концентрации металлов на разных этапах. Выявлено, что доля золота в концентрате постепенно увеличивается в течение всего технологического процесса, доля серебра в концентрате постепенно уменьшается, а содержание свинца незначительно увеличивается после первичной очистки, а затем особенно не меняется;
- произведено сравнение распределения размеров гранул сырья на обучающей и тестовой выборках. Каких-либо значимых различий в распределениях размеров гранул сырья не выявлено. Определено, что после обогащения исходного сырья происходило расщепление гранул;
- исследована суммарная концентрация веществ (золото, серебро, свинец, разные соли) на разных стадиях тех. процесса. Выявлено, что концентрация рассматриваемых веществ увеличилась в черновом концентрате после флотации. После первичной очистки суммарная концентрация веществ в среднем уменьшилась. Это может объясняться добавлением значительного числа реагентов. После вторичной очистки концентрация веществ снова увеличилась, а также уменьшился разброс значений, что говорит об успешности тех. процесса. Выявлено наличие нулевых значений суммарной концентрации веществ. Соответствующие объекты удалены из всех датасетов, так как они могут исказить предсказания модели.

Далее выполнен этап построения различных моделей и подсчёта требуемых метрик качества:

- произведена подготовка признаков к построению модели: выделены обычные и целевые признаки, выполнено масштабирование признаков;
- построена констатная модель, результаты работы которой учтены до построения реальных моделей;
- построены модели линейной регресии, "Дерево решений" и "Случайный лес". Для последних двух моделей подобраны гиперпараметры. По результатам расчёта метрик качества sMAPE для решения задачи выбрана модель "Случайный лес";
- проверка выбранной модели на тестовой выборке показала итоговое значение менее 8% для метрики sMAPE, которая учитывает эффективности обогащения чернового и финального концентратов.
