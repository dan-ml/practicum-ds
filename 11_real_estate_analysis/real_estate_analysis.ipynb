## Предобработка данных

### Заполнение пропусков

Первый столбец, в котором замечены пропуска, - ceiling_height. В столбце также были замечены редкие странные значения (к примеру, высота потолка 1 м). От них избавимся в дальнейшем в разделе 4.3. Пропусков в столбце очень много, около 38 процентов от общего числа значений. Изначально я заполнил пропуски медианным значением, но в дальнейшем в п. 4.1 понял, что это привело к серьёзному искажению данных и сдвинуло "усы" на диаграмме размаха. Поэтому от идеи заполнения пропусков медианным значением пришлось отказаться. Заполнить пропуски осмысленным значением не представляется возможным, значение высоты потолка нельзя вычислить на основе данных из других столбцов. Была идея связать высоту потолка с ценой, но и от неё пришлось отказаться. У этих двух величин крайне низкий коэффициент корреляции. Поэтому примем решение оставить пропуски в столбце ceiling_height и ориентироваться на те данные, которые в столбце присутствуют.

```python
main_data['ceiling_height'].corr(main_data['last_price'])
```

Рассмотрим на столбец floors_total. Количество этажей в доме может сильно отличаться для разных объектов. Предугадать верные значения нельзя. Медианными значениями заполнить прпоуски нельзя, а то может так случиться, что квартира на 10-ом этаже, а всего этажей 8. В данном столбце лучше оставить пропуски как есть.

Следующий столбец, в котором присутствуют пропуски, - living_area. Пропуски жилой площади можно попробовать заполнить вычисленными значениями на основе данных из других столбцов. В таблице присутствуют все значения общей площади квартир (total_area). Можно посчитать какую часть в среднем (среднее медианное значение) занимает жилая площадь от общей площади квартиры (пусть это будет переменная *part*). Далее пропуски жилой площади заполнить по формуле: `total_area * part`. Результат округлить до двух цифр после запятой. Затем сразу посмотрим правильность выполненных преобразований.

```python
part = (main_data['living_area']/main_data['total_area']).median()
main_data.loc[main_data['living_area'].isna(),'living_area'] = round(
    (main_data.loc[main_data['living_area'].isna(),'total_area'] * part), 2
)
main_data.head(10)
```

```python
main_data['living_area'].isna().sum()
```

На примере строк 3 и 7 видна корректная замена пропусков. По результатам команды isna() я точно убедился, что пропусков в столбце больше нет.

Следующий столбец с пропусками - is apartment. Посмотрим его уникальные значения командой unique(). 

```python
main_data['is_apartment'].unique()
```

Столбец имеет тип данных object, что не совсем корректно. Должен быть булевый тип (bool). Это исправим в следующем разделе. А сейчас заполним пропуски в этом столбце булевым значением False. Если для квартиры не указано, относится она к апартаментам или нет, то вероятнее всего, это жилая квартира. Для проверки правильности заполнения снова вызовем метод unique().

```python
main_data['is_apartment'] = main_data['is_apartment'].fillna(False)
main_data['is_apartment'].unique()
```

После заполнения пропусков тип данных в столбце *is_apartment* автоматически стал верным. Можно в этом убедиться, напрямую посмотрев тип объекта Series. Преобразование типа данных в нём уже не потребуется.

```python
main_data.is_apartment.dtype
```

Следующий столбец с пропусками - *kitchen_area*. Заполним пропуски по аналогии с жилой площадью. Посмотрим какую часть занимает площадь кухни от общей площади квартиры (для всех квартир), а затем найдем медианное значение этой части. На основе этого значения посчитаем площадь кухни для всех квартир, где она не указана, и заполним пропуски.

```python
part = (main_data['kitchen_area']/main_data['total_area']).median()
main_data.loc[main_data['kitchen_area'].isna(),'kitchen_area'] = round(
    (main_data.loc[main_data['kitchen_area'].isna(),'total_area'] * part), 2
)
main_data.head(5)
```

```python
main_data['kitchen_area'].isna().sum()
```

По строке №3 можно видеть, что пропуск заполнился адекватным значением. И в столбце точно не осталось других пропусков. 

Следующий столбец с пропусками - balcony. Посмотрим его уникальные значения.

```python
main_data['balcony'].unique()
```

Виден сразу неподходящий тип данных, количество балконов не должно быть вещественным. Это исправим в пункте 2.2. А сейчас заполним пропуски. Медианным значением (округлённым до целых) заполнять нелогично, так как число балконов пользователь мог специально не указывать по причине их отсутствия. Поэтому заполним пропуски в этом столбце нулями.

```python
main_data['balcony'] = main_data['balcony'].fillna(0)
main_data['balcony'].unique()
```

Заполнение пропусков прошло успешно.

Следующий столбец с пропусками - locality_name. Сразу отмечу, что пропусков в этом столбце очень мало (около 0,2% от общего числа строк). Значения столбца важные, а заполнить их корректно нельзя. Поэтому лучше удалить все строки, имеющие пропуски в столбце locality_name. Также нужно исправить нумерацию индексов. После удаления строк сразу проверим оставшееся их количество в таблице.

```python
main_data = main_data.dropna(subset=['locality_name']).reset_index(drop=True)
len(main_data)
```

Удаление строк прошло успешно.

Следующий столбец с пропусками - airports_nearest. Если для какой-то квартиры в этом столбце присутствует пропуск, то скорее всего рассматривается какой-то населённый пункт, в котором аэропорта нет. Если же пропуск присутствует у квартир из крупного города с аэропортом, то нет возможности заполнить его каким-то осмысленным значением. При этом заполнять пропуски в столбце нулевыми значениями нелогично. Поэтому примем решение оставить пропуски как есть. Аналогично поступим со стобцами parks_nearest и pond_nearest. Если городок небольшой, в нём может не быть ни парка, ни водоёма.

Пропуски присутствуют ещё в столбце cityCenters_nearest. Этот столбец тоже актуален только для крупного города, к примеру, для Санкт-Петербурга. В мелких городках или посёлках понятие "центр города" либо отсутствует, либо оно условное и не подразумевает какой-то большой район. Примем решение оставить в этом столбце пропуски как есть.

Следующие столбцы с пропусками - parks_around3000 и ponds_around3000. Они не будут использоваться далее в анализе данных, поэтому оставим в них пропуски как есть. Информация о наличии пропусков в столбцах может помочь коллегам лучше подготавливать исходные данные для анализа.

Последний столбец с пропусками - days_exposition. Его значения пригодятся в дальнейшем при анализе. Прежде чем принять решение о возможности заполнения пропусков в этом столбце, посмотрим его связь со столбцом first_day_exposition, используя диаграмму рассеяния. Так получится увидеть, как значения столбца days_exposition менялись во времени. 
Перед построением диаграммы необходимо преобразовать тип столбца first_day_exposition из object в datetime.

```python
main_data['first_day_exposition'] = pd.to_datetime(main_data['first_day_exposition'], format='%Y-%m-%d')
main_data.plot(
    x='first_day_exposition', y='days_exposition', title='Зависимость срока продажи от даты публикации объявления', 
    kind='scatter', figsize=(10,6), alpha=0.1)
plt.show()
```

По диаграмме рассеяния нельзя ничего сказать о причинах пропусков. Но можно сделать вывод, что заполнять пропуски медианными значениями нельзя. Сроки продажи квартир (количество дней нахождения объявлений в состоянии "опубликовано") не распределены случайно, относительно дат публикации объявлений. По диаграмме видно, что начиная с 2015 года минимальный срок продажи значительно сократился и с середины 2016-ого года квартиры стали продаваться быстро, в том числе за срок меньше 1 недели. Причины могут быть следующие:
- на сайте с объявлениями стали появляться мошеннические объявления, которые потом быстро снимались с публикации;
- часть квартир ещё находится в состоянии продажи и объявления пока с продажи не сняты (это частично может объяснить наличие пропусков в столбце days_exposition);
- засчёт развития сервиса Яндекс.Недвижимость продажи действительно стали происходить быстрее.

Исходя из вышесказанного примем решение оставить пропуски в столбце days_exposition как есть.

### Преобразование типов данных

Как я уже описывал ранее, столбец is_apartment имел некорректный тип object, но после заполнения пропусков тип столбца автоматически исправился на булевый. Так что теперь преобразовывать тип этого столбца не надо. 

Столбец first_day_exposition в п 2.1 уже был преобразован в datetime.

Обратим внимание на перечень столбцов, имеющих вещественных тип, который следует преобразовать в целочисленный: last_price, floors_total, balcony, parks_around3000, ponds_around3000, days_exposition. В столбце last_price не имеет практического смысла оставлять значения десятичных долей денежных единиц, так как суммы огромные и долями можно пренебречь. Остальные перечисленные мной столбцы содержат количественные значения чего-то (этажи, балконы, парки и пр.), которые должны быть целыми. Столбец last_price перед преобразованием в целочисленный тип стоит  перевести в более крупные денежные единицы. К примеру, если столбец хранит значения в рублях, то будет хранить значения в тыс. руб. С такими числами далее будет проще работать. 

Преобразуем типы столбцов методами astype или convert_dtypes. Для столбцов, содержащих значения NaN, следует использовать метод convert_dtypes с прямым указанием на конвертацию в целочисленный тип. Вещественные значения NaN заменятся на значения NA, которые допустимы для целочисленного формата данных.

Изменять тип данных столбца days_exposition сейчас не будем, так как в п. 4.3 к значениям этого столбца будем применяться функция np.nanpercentile, которая требует наличия именно значений np.NaN, а не pd.NA.

Затем посмотрим заново общую информацию о таблице.


```python
main_data['last_price'] = (main_data['last_price']/1000).astype('int')
main_data['floors_total'] = main_data['floors_total'].convert_dtypes(convert_integer=True)
main_data['balcony'] = main_data['balcony'].astype('int')
main_data['parks_around3000'] = main_data['parks_around3000'].convert_dtypes(convert_integer=True)
main_data['ponds_around3000'] = main_data['ponds_around3000'].convert_dtypes(convert_integer=True)
main_data.info()
```

Преобразования типов прошли корректно. 

## Расчёты и добавление результатов в таблицу

### Добавление столбца с типом этажа

Добавим столбец floor_group с типом этажа квартиры. Для этого напишем функцию, которая определяет группу в зависимости от этажа - первый, последний или другой. Проверим работу функции. Так как в столбце floors_total целенаправленно оставлены значения NA, то проверим работу функции и с ними. Затем применим функцию к столбцу floor и непосредственно создадим столбец floor_group. 

Чтобы вызов созданной функции не приводил к ошибке при наличии в строке значения NA, в теле функции проверим один из элементов строки на это значение. В случае присутствия значения NA завершим функцию с результатом "другой".

В связи с тем, что в п. 4.4 будут считаться коэффициенты корреляции между различными параметрами (в том числе и с типом этажа), нельзя использовать строковые значения для столбца floor_group. Заменим строковые значения на номера, которые и будут возвращаться функцией:
- 1 - первый этаж;
- 2 - другой этаж (ни первый, ни последний, либо номер этажа неизвестен).;
- 3 - последний этаж

```python
def floor_group(row):
    if row.floor==1:
        return 1
    elif pd.isna(row.floors_total): 
        return 2
    elif row.floor==row.floors_total:
        return 3
    else:
        return 2

display(floor_group(pd.Series({'floor': 1, 'floors_total': 10})))
display(floor_group(pd.Series({'floor': 1, 'floors_total': 1})))
display(floor_group(pd.Series({'floor': 2, 'floors_total': 12})))
display(floor_group(pd.Series({'floor': 8, 'floors_total': 8})))
display(floor_group(pd.Series({'floor': 4, 'floors_total': None})))

main_data['floor_group'] = main_data[['floor', 'floors_total']].apply(floor_group, axis=1)
main_data[['floors_total', 'floor', 'floor_group']].head(5)
```

Функция сработала верно. Посмотрим, верно ли добавились значения нового столбца для строк, где количество этажей равно NA.

```python
main_data[main_data['floors_total'].isna()].head(5)
```

### Исключение редких и выбивающихся значений

Уберём из датасета строки, которые содержат редкие и выбивающиеся значения в исследованных выше пяти столбцах: площадь, цена, число комнат, высота потолков, время продажи. 

Выбросами будут считаться значения, которые находятся за пределеами "усов" (1,5\*IQR от границ "ящика" в обе стороны). Напишем функцию, которая будет вычислять для каждого исследуемого столбца нижнюю и верхнюю границу отсева. В функции будут использоваться значения квартилей Q1 и Q3, рассчитанные методами `percentile` или `nanpercentile` из библиотеки numpy. Проверим работу функции, указав в качестве её аргумента столбец days_exposition, а также столбец ceiling_height, содержащий значения NaN.

```python
def get_bounds(data):
    if data.isna().sum()==0:
        q75,q25 = np.percentile(data,[75,25])
    else:
        q75,q25 = np.nanpercentile(data,[75,25])
    iqr = q75-q25
    max = q75+(1.5*iqr)
    min = q25-(1.5*iqr)
    if max>data.max():
        max=data.max()
    if min<data.min():
        min=data.min()
    return min, max

display(get_bounds(main_data['days_exposition']))
display(get_bounds(main_data['ceiling_height']))
```

Функция работает верно. Теперь используем созданную функцию, чтобы находить границы до выбросов данных в исследуемых столбцах. Будем последовательно формировать срезы данных из исходной таблицы на основе выходных данных функции get_bounds. Исходный датасет не будем изменять, для дальнейшей работы создадим датасет main_data_good. При рассмотрении столбцов ceiling_height и days_exposition в срез добавим также значения NaN, которые я ранее решил не заполнять.

```python
min_bound, max_bound = get_bounds(main_data['total_area'])
main_data_good = main_data.query('total_area >= @min_bound and total_area <= @max_bound')

min_bound, max_bound = get_bounds(main_data['last_price'])
main_data_good = main_data_good.query('last_price >= @min_bound and last_price <= @max_bound')

min_bound, max_bound = get_bounds(main_data['rooms'])
main_data_good = main_data_good.query('rooms >= @min_bound and rooms <= @max_bound')

min_bound, max_bound = get_bounds(main_data['ceiling_height'])
main_data_good = main_data_good[((main_data_good['ceiling_height']>=min_bound) & 
                                 (main_data_good['ceiling_height']<=max_bound) | 
                                 (main_data_good['ceiling_height'].isna()))]

min_bound, max_bound = get_bounds(main_data['days_exposition'])
main_data_good = main_data_good[((main_data_good['days_exposition']>=min_bound) & 
                                 (main_data_good['days_exposition']<=max_bound) | 
                                 (main_data_good['days_exposition'].isna()))]

main_data_good = main_data_good.reset_index(drop=True)
main_data_good.shape
```

Так как функция np.nanpercentile к столбцу days_exposition уже была применена, можно преобразовать тип этого столбца в целочисленный. Но простым применением метода convert_dtypes обойтись не получится, ровно как и функцией int(). В столбце days_exposition содержатся как значения c дробной частью (нечто типа 45.00000003), так и оставленные пропуски. В цикле округлим значения столбца до целых, а затем уже применим метод convert_dtypes.

```python
for i in range(main_data_good['days_exposition'].shape[0]):
    if not pd.isna(main_data_good.iloc[i, 20]):
        main_data_good.iloc[i, 20] = round(main_data_good.iloc[i,20],0)
main_data_good['days_exposition'] = main_data_good['days_exposition'].convert_dtypes(convert_integer=True)
main_data_good['days_exposition'].head(5)
```

По поводу двух найденных в предыдущем разделе аномальных значений 45 и 60 в столбце days_exposition: удалять строки с данными значениями неправильно, так как могут быть реальные продажи, которые заняли 45 или 60 дней. При этом невозможно предположить, какие строки с данными во всей таблице соответствуют реальным продажам в 45 и 60 дней, а какие - к ошибочным. Следует обсудить этот вопрос с сотрудниками, занимающимися подготовкой данных.

### Определение области, считающейся центром г. Санкт-Петербурга

Создадим таблицу со срезом данных, включающую только объявления для г. Санкт-Петербурга. Посмотрим информацию о получившейся таблице.

```python
data_good_spb = main_data_good.query('locality_name == "Санкт-Петербург"')
data_good_spb.info()
```

Нас интересует столбец city_centers_nearest, а он содержит небольшое число пропусков. Избавимся от строк с пропусками в этом столбце.

```python
data_good_spb = data_good_spb.dropna(subset=['city_centers_nearest']).reset_index(drop=True)
data_good_spb.shape
```

Удаление строк прошло успешно. Теперь создадим в таблице столбец city_centers_km с расстоянием до центра в километрах. Сразу преобразуем тип данных столбца в целочисленный. Перед преобразованием округлим числа до целых значений по правилам математики функцией `round`.

```python
data_good_spb['city_centers_km'] = (round(data_good_spb['city_centers_nearest']/1000,0)).astype('int')
data_good_spb[['city_centers_nearest', 'city_centers_km']].head(5)
```

Добавление столбца и преобразование типа данных прошли успешно. Для определения средней стоимости жилья для каждого километра создадим сводную таблицу, в которой первичная группировка значений будет произведена по столбцу city_centers_km. Для созданной таблицы сразу построим график. Для наглядности включим отображение промежуточных линий сетки.

```python
(data_good_spb.pivot_table(index='city_centers_km', values='last_price').
    plot(title='Средняя стоимость каждого км от центра', figsize=(10,6), grid=True, xlabel='Расстояние от центра', ylabel='Цена'))
plt.minorticks_on()
plt.grid(which='minor', linestyle = ':')
plt.show()
```

**Вывод:** центром города при определении стоимости жилья можно считать область в радиусе 8 км от фактического центра города.
